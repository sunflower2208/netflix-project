```python
import pandas as pd
```

```python
imdb_title = pd.read_csv("datasets/title.basics.tsv.gz",sep='\t')
imdb_rating = pd.read_csv("datasets/title.ratings.tsv.gz",sep='\t')
imdb_crew = pd.read_csv("datasets/title.crew.tsv.gz",sep='\t')
imdb_names = pd.read_csv("datasets/name.basics.tsv.gz",sep='\t')
```

```python
imdb_title.head()
```

```python
imdb_title = imdb_title.drop(labels=['isAdult', 'endYear', 'runtimeMinutes', 'genres'], axis = 1)
```

```python
imdb_rating.head()
```

```python
imdb_crew.head()
```

```python
imdb_names.head()
```

```python
imdb_names = imdb_names.drop(labels=['birthYear', 'deathYear', 'primaryProfession', 'knownForTitles'], axis = 1)
```

```python
imdb_titles = pd.merge(imdb_title, imdb_rating, how = 'inner', on = 'tconst')
```

```python
imdb_titles.head()
```

```python
imdb_people = pd.merge(imdb_crew, imdb_names, how = 'inner', left_on = 'directors', right_on = 'nconst')
```

```python
imdb_people.head()
```

```python
imdb_people = imdb_people.drop(labels=['directors', 'writers', 'nconst'], axis = 1)
```

```python
imdb = pd.merge(imdb_titles, imdb_people, how = 'left', on = 'tconst')
imdb.head()
```

```python
imdb.info()
```

```python
imdb = imdb.drop(imdb.loc[(imdb['titleType'] == 'tvEpisode') | (imdb['titleType'] == 'videoGame')].index)
```

```python
imdb['titleType'].value_counts()
```

```python
imdb['titleType'] = imdb['titleType'].apply(lambda x: 'Movie' if x in ['short', 'movie', 'tvMovie', 'video', 'tvSpecial', 'tvShort'] else x)
imdb['titleType'] = imdb['titleType'].apply(lambda x: 'TV Show' if x in ['tvSeries', 'tvMiniSeries'] else x)
```

```python
imdb['titleType'].value_counts()
```

```python
imdb.drop('tconst', axis = 1).to_pickle("datasets/imdb.pkl")
```
